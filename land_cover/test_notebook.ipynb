```python
import ee
import geemap
```

```python
ee.Initialize()
startDate = '2015-01-01';
endDate = '2021-10-25';
geometry = ee.Geometry.Point([-43.2777477551557, -22.950154540990972]);
```

```python
Map = geemap.Map()
Map.add_basemap('HYBRID')
Map
```

```python
Map.addLayer(geometry, {'color': 'black'}, 'Selected Location');
Map.centerObject(geometry, 16)
```

```python
s2 = ee.ImageCollection('COPERNICUS/S2_HARMONIZED').filterDate(startDate, endDate).filterBounds(geometry);
```

```python
s2.first().getInfo()
```

```python
s2Image = ee.Image(s2.first());
```

```python
s2Image.getInfo()
```

```python
imageId = s2Image.get('system:index');
```

```python
imageId.getInfo()
```

```python
dw = ee.ImageCollection('GOOGLE/DYNAMICWORLD/V1').filterDate(startDate, endDate).filterBounds(geometry);
```

```python
dw.getInfo()
```

```python
dwImage = ee.Image(dw.first())
```

```python
dwImage.getInfo()
```

```python
classification = dwImage.select('label')
```

```python
classification
```

```python
classification.getInfo()
```

```python
dwImage.sample(geometry).first().get('label').getInfo()
```

```python
dwVisParams = {
  'min': 0,
  'max': 8,
  'palette': [
    '#419BDF', '#397D49', '#88B053', '#7A87C6', '#E49635', '#DFC35A',
    '#C4281B', '#A59B8F', '#B39FE1'
  ]
}
```

```python
Map.addLayer(classification, dwVisParams, 'Classified Image');
Map.add_legend(title="Dynamic World Land Cover", builtin_legend='Dynamic_World')
```
